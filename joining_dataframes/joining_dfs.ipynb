Import Libraries

```python
import pandas as pd
```

Load datasets

```python
df_cust = pd.read_csv("data/customer_data.csv")
df_trans = pd.read_csv("data/transaction_data.csv")
df_products = pd.read_csv("data/products_data.csv")
```

Verify the data

```python
df_cust
```

```python
df_trans
```

Inner Join

```python
pd.merge(df_cust, df_trans, on='Cust_ID', how='inner')
```

Left Join

```python
pd.merge(df_cust, df_trans, on='Cust_ID', how='left')
```

Right Join

```python
pd.merge(df_cust, df_trans, on='Cust_ID', how='right')
```

Outer Join

```python
pd.merge(df_cust, df_trans, on='Cust_ID', how='outer')
```

Joining 3 dataframes

```python
df_cust.merge(df_trans, on='Cust_ID').merge(df_products, on='Prod_ID')
```

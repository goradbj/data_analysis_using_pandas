Handling Missing Data

1. Drop Column : If column is irrelavent and has many missing data, we can drop it

```python

df.address.unique()
```

```python
df.address.nunique()
```

```python
df.address.value_counts()
```

```python
df = df.drop(columns=['address'], axis=1)
df.head()
```

2. Removing Rows: dropna() - one approach when we have very less missing values then we can drop those rows

```python
df1 = df.dropna()
df1
```

```python
df
```

3. Not dropping rows, fill missing datas for categorical columns with mode

Gender

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.gender.value_counts()
```

```python
df.gender.mode()[0]
```

```python
df.gender = df.gender.fillna(df.gender.mode()[0])
df.isna().sum()
```

smoking_history

```python
df.smoking_history.value_counts()
```

```python
df.smoking_history = df.smoking_history.fillna("No Info")
df.isna().sum()
```

4. Not dropping rows, fill missing datas for numerical columns (Age) using Mean

```python
df.age.mean()
```

```python
df.age = df.age.fillna(int(df.age.mean()))
df.isna().sum()
```

5. Not dropping rows, fill missing datas for numerical columns using Mode

```python
df.hypertension.value_counts()
```

```python
df.hypertension.mode()
```

```python
df.hypertension = df.hypertension.fillna(df.hypertension.mode()[0])
df.isna().sum()
```

```python
df.heart_disease = df.heart_disease.fillna(df.heart_disease.mode()[0])
df.isna().sum()
```

6. Forward Fill

```python
df.HbA1c_level.isna()
```

```python
df
```

```python
df.HbA1c_level = df.HbA1c_level.fillna(method='ffill')
df
```

```python
df.isna().sum()
```

7. Backward Fill

```python
df.blood_glucose_level.isna()
```

```python
df.blood_glucose_level = df.blood_glucose_level.fillna(method='bfill')
df
```

```python
df.blood_glucose_level = df.blood_glucose_level.fillna(method='ffill')
df
```

8. Drop Rows which have NA

```python
df = df.dropna()
df
```

8. Drop duplicates

```python
df = df.drop_duplicates()
df
```
